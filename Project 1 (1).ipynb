# Downloading Fasta Sequences

```python
L1_sseqid=L1_df["sseqid"]
L1_sseqid_list=list(L1_sseqid)
print(L1_sseqid_list)
```

```python
from Bio import Entrez, SeqIO

# Provide your email address for NCBI
Entrez.email = "blkoger@uw.edu"

# Use efetch to download the DNA sequences
handle = Entrez.efetch(db="nucleotide", id=L1_sseqid_list, rettype="fasta", retmode="text")
records = SeqIO.parse(handle, "fasta")

# Write the downloaded sequences to a file
with open("blaL1_dna_seqs", "w") as output:
    SeqIO.write(records, output, "fasta")

# Close the handle to free up memory
handle.close()

print("We are done!")
```

```python
from Bio import SeqIO
from Bio import Entrez

output_file = "blaL1_100_seqs_parsed_2.fasta"  # Output file name

# Provide your email address for NCBI Entrez
Entrez.email = "blkoger@uw.edu"

with open(output_file, "w") as f:
    for index, row in L1_df.iterrows():
        # Extract the sequence ID, start position, and end position from the columns
        seq_id = row["sseqid"]
        start_pos = int(row["sstart"])
        end_pos = int(row["send"])

        try:
            # Fetch the sequence from NCBI using the provided sequence ID
            handle = Entrez.efetch(db="nucleotide", id=seq_id, rettype="fasta", retmode="text")
            record = SeqIO.read(handle, "fasta")
            handle.close()

            # Extract the sequence for the specified region
            subsequence = record.seq[start_pos:end_pos]

            # Write the sequence to the output file
            f.write(f">{seq_id}_{start_pos}-{end_pos}\n{subsequence}\n")
        except Exception as e:
            # Log the error and continue to the next sequence
            print(f"Error occurred for sequence ID: {seq_id}. Skipping. Error message: {str(e)}")
```
